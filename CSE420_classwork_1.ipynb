```python
from PIL import Image
import matplotlib.pyplot as plt
import os

img_path = '/content/drive/My Drive/flower.png'


im = Image.open(img_path)

plt.axis('off')
plt.imshow(im)
plt.show()
```

```python
import cv2
im = cv2.imread(img_path)
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.axis('off')
plt.show()
```

```python
import numpy as np
img_array = np.array(im)
img_part = img_array[0:100, 0:100]

print(img_array.shape)
print("red pixel values", img_array[360][360][0])
print("green pixel values", img_array[360][360][1])
print("blue pixel values", img_array[360][360][2])
```

```python
r, g, b = cv2.split(im)
plt.imshow(r, cmap='Reds')
plt.title('Red Channel')
plt.axis('off')
plt.show()

plt.imshow(r, cmap='Greens')
plt.title('Green Channel')
plt.axis('off')
plt.show()

plt.imshow(r, cmap='Blues')
plt.title('Blue Channel')
plt.axis('off')
plt.show()
```
